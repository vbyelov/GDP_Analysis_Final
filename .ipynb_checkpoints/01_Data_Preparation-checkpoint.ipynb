# Import Libraries & Load Dataset <a id="import-libraries"></a>

Data Import & Initial Inspection <a id="data-import"></a>

**Goal:** Set up the notebook, define the data path, and prepare to load the raw CSV.

**File layout:** `daten/GDP_2005_2025_final.csv`

**Actions in the next code cell:**
- Import `pandas` and `numpy`
- Set `DATA_PATH = "daten/GDP_2005_2025_final.csv"`
- Plan to check:
  - `shape`
  - `columns`
  - `dtypes`

**Expected output:**  
Basic structure info ready for data cleaning.


```python
# Step 1 ‚Äî Import Libraries & Load Dataset

import pandas as pd
import numpy as np

# Define the data path
DATA_PATH = "daten/GDP_2005_2025_final.csv"

# Load the dataset
Dataset = pd.read_csv(DATA_PATH)

# Basic info
print("Shape:", Dataset.shape)
print("\nColumns:", Dataset.columns.tolist())
print("\nData types:\n", Dataset.dtypes.head())
```

# Initial Data Inspection <a id="data-inspection"></a>


**Goal:** Understand data quality and completeness before cleaning.

**Actions in the next code cell:**
- Display first few rows using `head()`
- Check for missing values with `isna().sum()`
- Review basic statistics with `describe()`

**Expected output:**
- A quick visual overview of the dataset
- Insight into possible missing or inconsistent data

```python
# Step 2 ‚Äî Initial Data Inspection

# Display the first 5 rows
pd.set_option("display.float_format", "{:,.0f}".format)
display(Dataset.head())

# Check for missing values
print("\nMissing values per column:\n")
print(Dataset.isna().sum())

# Basic descriptive statistics
print("\nDescriptive statistics:\n")
display(Dataset.describe())
```

# Data Cleaning & Transformation <a id="data-cleaning"></a>


**Goal:** Ensure consistent data before reshaping for analysis.

**Actions in the next code cell:**
- Confirm all numeric columns are of type `float64`
- Check if any countries have missing values across all years
- Decide how to handle missing values (fill or keep)
- Verify there are no duplicated country names

**Expected output:**
- Clean, consistent dataset ready for transformation


```python
# Step 3 ‚Äî Data Cleaning

# Format numbers for better readability
pd.set_option("display.float_format", "{:,.0f}".format)

# Ensure all numeric columns are float
print("Data types check:\n")
print(Dataset.dtypes.value_counts())

# Check for countries with all NaN values (just in case)
print("\nCountries with all NaN values across years:\n")
empty_rows = Dataset[Dataset.iloc[:, 1:].isna().all(axis=1)]
display(empty_rows[["Country"]])

# Check for duplicated country names
print("\nDuplicated country names:")
print(Dataset["Country"].duplicated().sum())

# Decide how to handle missing values (here we keep them for now)
print("\nMissing values summary (final check):")
print(Dataset.isna().sum().tail())
```

# Reshape Data for Analysis <a id="reshape-data"></a>

**Goal:** Convert the dataset from wide to long format for time series analysis and visualization.

**Actions in the next code cell:**
- Use `pd.melt()` to transform columns (years) into rows
- Rename columns to `Country`, `Year`, and `GDP`
- Convert `Year` to integer type
- Check resulting structure with `head()` and `info()`

**Expected output:**
A tidy dataset with three columns: `Country`, `Year`, `GDP`.


```python
# Step 4 ‚Äî Reshape Data for Analysis

# Melt wide ‚Üí long
year_cols = [c for c in Dataset.columns if c != "Country"]
Dataset_long = pd.melt(
    Dataset,
    id_vars=["Country"],
    value_vars=year_cols,
    var_name="Year",
    value_name="GDP"
)

# Types & tidy
Dataset_long["Year"] = Dataset_long["Year"].astype(int)
Dataset_long["GDP"] = pd.to_numeric(Dataset_long["GDP"], errors="coerce")

# Quick checks
print("Shape long:", Dataset_long.shape)
display(Dataset_long.head(10))
Dataset_long.info()
print("\nMissing GDP values:", Dataset_long["GDP"].isna().sum())
```

# Descriptive Statistics and Grouped Analysis <a id="descriptive-stats"></a>


**Goal:** Explore GDP distribution and compare countries and years.

**Actions in the next code cell:**
- Compute overall min, max, mean GDP
- Find top 5 and bottom 5 GDP values for 2020
- Calculate average GDP per year across all countries
- Visualize quick trend (line chart) with `plot()`

**Expected output:**
Key insights into GDP scale and evolution over time.


```python
# Step 5 ‚Äî Descriptive Statistics and Grouped Analysis

# Basic stats
print("Overall GDP statistics:\n")
print(Dataset_long["GDP"].describe())

# Top and bottom 5 GDP countries in 2020
print("\nTop 5 countries by GDP in 2020:")
top5_2020 = (
    Dataset_long[Dataset_long["Year"] == 2020]
    .sort_values(by="GDP", ascending=False)
    .head(5)
)
display(top5_2020)

print("\nBottom 5 countries by GDP in 2020:")
bottom5_2020 = (
    Dataset_long[Dataset_long["Year"] == 2020]
    .sort_values(by="GDP", ascending=True)
    .head(5)
)
display(bottom5_2020)

# Average GDP per year across all countries
avg_gdp_year = Dataset_long.groupby("Year")["GDP"].mean()

# Quick visualization
ax = avg_gdp_year.plot(title="Average Global GDP over Time", figsize=(8,4))
ax.set_xticks(avg_gdp_year.index)      
ax.set_xticklabels(avg_gdp_year.index, rotation=45)
ax.set_xlabel("Year")
ax.set_ylabel("Average GDP")
```

# Country-Level GDP Trends <a id="country-trends"></a>


**Goal:** Compare GDP growth patterns across selected countries.

**Actions in the next code cell:**
- Select a few countries (e.g., United States, China, Germany, Japan, India)
- Plot their GDP trends over time
- Use clear labels and legend
- Optionally add gridlines for readability

**Expected output:**
A multi-line chart showing GDP evolution of major economies.


```python
# Step 6 ‚Äî Country-Level GDP Trends

import matplotlib.pyplot as plt

# Pick countries (change if you want)
countries = ["United States", "China", "Germany", "Japan", "India"]

# Filter and pivot to Year √ó Country
df_sel = Dataset_long[Dataset_long["Country"].isin(countries)].copy()
pivot = df_sel.pivot(index="Year", columns="Country", values="GDP").sort_index()

# Plot
ax = pivot.plot(figsize=(9,5), title="GDP Trends: Selected Countries (2005‚Äì2025)")
ax.set_xticks(pivot.index)
ax.set_xticklabels(pivot.index, rotation=45)
ax.set_xlabel("Year")
ax.set_ylabel("GDP")
ax.grid(True)
plt.tight_layout()
```

# GDP Growth Comparison <a id="gdp-growth"></a>


**Goal:** Analyze GDP growth rate between two years to identify fastest-growing economies.

**Actions in the next code cell:**
- Calculate GDP growth rate between 2005 and 2025
- Sort countries by growth percentage
- Display top 10 and bottom 10 countries
- Visualize with a horizontal bar chart

**Expected output:**
Ranking of countries by relative GDP growth over the period.


```python
# Step 7 ‚Äî GDP Growth Comparison (2005 ‚Üí 2025)

import matplotlib.pyplot as plt

# Select baseline & target years
baseline_year = "2005"
target_year = "2025"

# Prepare growth frame
growth_df = (
    Dataset[["Country", baseline_year, target_year]]
    .rename(columns={baseline_year: "GDP_2005", target_year: "GDP_2025"})
    .dropna(subset=["GDP_2005", "GDP_2025"])
)

# Avoid division by zero
growth_df = growth_df[growth_df["GDP_2005"] != 0]

# Compute growth
growth_df["Growth_abs"] = growth_df["GDP_2025"] - growth_df["GDP_2005"]
growth_df["Growth_pct"] = growth_df["Growth_abs"] / growth_df["GDP_2005"]

# Rankings
ranked = growth_df.sort_values("Growth_pct", ascending=False).reset_index(drop=True)

# Show top 10 / bottom 10
print("Top 10 countries by GDP growth rate (2005‚Üí2025):\n")
display(
    ranked[["Country", "GDP_2005", "GDP_2025", "Growth_abs", "Growth_pct"]]
    .head(10)
    .assign(Growth_pct=lambda d: (d["Growth_pct"]*100).round(1))
)

print("\nBottom 10 countries by GDP growth rate (2005‚Üí2025):\n")
display(
    ranked[["Country", "GDP_2005", "GDP_2025", "Growth_abs", "Growth_pct"]]
    .tail(10)
    .assign(Growth_pct=lambda d: (d["Growth_pct"]*100).round(1))
)

# Visualization: Top 10 growth rates
top10 = ranked.head(10).copy()
top10 = top10.sort_values("Growth_pct", ascending=True)  # for a nice bottom‚Üítop barh

ax = top10.plot(
    x="Country", y="Growth_pct", kind="barh",
    figsize=(8,5), legend=False,
    title="Top 10 GDP Growth Rates (2005‚Üí2025)"
)
ax.set_xlabel("Growth rate")
ax.set_ylabel("")
ax.xaxis.set_major_formatter(lambda v, pos: f"{v*100:.0f}%")
ax.grid(True, axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
```

# Correlations and Distributions <a id="correlations"></a>


**Goal:** Examine how GDP values correlate across time and understand distribution patterns.

**Actions in the next code cell:**
- Compute correlation matrix between years
- Visualize it with a heatmap
- Plot distribution (histogram) of GDP values (log scale if needed)

**Expected output:**
Heatmap showing temporal correlations and histogram revealing GDP distribution.


```python
# Step 8 ‚Äî Correlations and Distributions

import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix between years (wide format works best)
corr = Dataset.drop(columns=["Country"]).corr()
#Heatmap
subset = Dataset[["2005", "2010", "2015", "2020", "2025"]]
plt.figure(figsize=(6,4))
sns.heatmap(subset.corr(), annot=True, cmap="vlag", center=0)
plt.title("GDP Correlation: 2005‚Äì2025 (every 5 years)")
plt.tight_layout()
plt.show()

# Histogram of all GDP values
plt.figure(figsize=(8,4))
sns.histplot(Dataset_long["GDP"].dropna(), bins=50, kde=True)
plt.title("Distribution of GDP Values (linear scale)")
plt.xlabel("GDP")
plt.tight_layout()
plt.show()

# Histogram with log scale
plt.figure(figsize=(8,4))
sns.histplot(np.log10(Dataset_long["GDP"].dropna()), bins=50, kde=True)
plt.title("Distribution of GDP Values (log10 scale)")
plt.xlabel("log10(GDP)")
plt.tight_layout()
plt.show()
```
